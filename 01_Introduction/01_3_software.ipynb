## 1.3 Software

To use Jupyter notebooks you need:

- CentOS 6 or later or some other recent Linux distribution, macOS 10.13 or later, or Windows 8 or later
- 5&nbsp;GB available on your disk
- an up-to-date version of Chrome, Firefox or Safari,
  as notebooks are not guaranteed to work well with other browsers.

We will use Anaconda, a free software collection for Linux, Mac and Windows
that includes Python, Jupyter and other libraries used in M269.
To install Anaconda, do the following:

1. Download the Individual Edition of Anaconda with Python 3.8 or later from
   https://www.anaconda.com/products/individual.
   To find out whether you have a 32- or 64-bit Windows version,
   click the Start button, select Settings > System > About, and then,
   under 'Device specifications', see 'System type'.
2. Check any other system requirements at
   https://docs.anaconda.com/anaconda/install and
   follow the installation instructions. Your operating system may warn you
   that the installer requires access to the downloads folder. That's fine.
   You can skip the optional step to install PyCharm: it's not needed in M269.
3. If you have problems during installation,
   see https://docs.anaconda.com/anaconda/user-guide/troubleshooting.

Next, you must put the `custom.css` file that styles notebooks in its right
place. The file is currently in the `book` folder that was extracted from
the archive you downloaded from the [M269 website](https://learn2.open.ac.uk/course/view.php?id=208089).

You can do the next steps using the graphical interface of your file explorer.
On macOS, you must press Cmd-Shift-. in the Finder before step&nbsp;4 and
after step&nbsp;6 in order to show (and then hide again) all files and folders
with names starting with a dot.

4. Create a folder `.jupyter` (note the dot) in your home folder.
   If your username is `me`, then your home folder is usually
   `/home/me` on Linux, `/Users/me` on macOS, and `C:\Users\me` on Windows.
5. Create a subfolder of `.jupyter` named `custom`.
6. Move or copy the `custom.css` file from the `book` folder to the `custom`
   folder. (I prefer to copy in order to keep a backup in the `book` folder.)

Alternatively, do steps 4–6 with the command line, as follows.

- If you have Linux or macOS:
    4. Open a terminal.
    5. Enter `mkdir -p ~/.jupyter/custom`.
    6. Enter `cp .../book/custom.css ~/.jupyter/custom`,
       replacing `...` with the path to your M269 folder, e.g.
       `cp ~/Documents/M269/book/custom.css ~/.jupyter/custom`.
- If you have Windows:
    4. Open the Command Prompt.
    5. Enter `mkdir .jupyter`, then `mkdir .jupyter\custom`.
    6. Enter `copy ...\book\custom.css .jupyter\custom`,
       replacing `...` with the path to your M269 folder, e.g.
       `copy C:\Users\me\Documents\M269\book\custom.css .jupyter\custom`.

Now you can start using Jupyter.

- If you use Linux or macOS:
    7. Open a terminal (or use the one you opened in step&nbsp;4 above),
       and enter `jupyter notebook`.
- If you use Windows:
    7. Run the application Anaconda3 > Jupyter Notebook.

Your operating system may ask which browser you want to use.

As the last two steps show, every time you click on a link to a notebook,
it's opened in a new tab. Unfortunately, if the notebook is already open,
the browser will open another copy of that notebook.
It's therefore easy to end up with lots of tabs, several of them showing the
same notebook. Make sure you close notebooks that you don't need anymore.
(I'll tell you how to properly close a notebook later.)
Avoid clicking on the 'previous section' link at the end of each notebook:
you probably have it already open.
You can easily see which sections you opened from the tab titles:
they start with the chapter and section number.

Now start working through the notebook you just opened.
